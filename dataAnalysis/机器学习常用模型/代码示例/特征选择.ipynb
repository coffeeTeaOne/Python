```python
from sklearn.datasets import load_iris
# 基于方差的选择
from sklearn.feature_selection import VarianceThreshold
```

```python
data = load_iris().data
target = load_iris().target
```

```python
data.std(axis=0)
```

```python
# threshold是方差的阈值，小于这个阈值的列将会被过滤掉
# 选择数据，就是选择方差大的数据，因为方差大的列保有更多的信息
VarianceThreshold(threshold=0.5).fit_transform(data)
```

```python
# 递归特征消除法
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression

logistic = LogisticRegression()
# 根据指定的算法模型，进行多次模型训练，每一次把系数最小的特征剔除，直到剔除到保留的个数为止
# n_features_to_select要保留的特征个数
RFE(logistic,n_features_to_select=2).fit_transform(data,target)
```

```python
from sklearn.feature_selection import SelectFromModel
from sklearn.linear_model import LogisticRegression

#带L1惩罚项的逻辑回归作为基模型的特征选择
# Logistic惩罚包括L1,L2，默认值是L2，
# 最终选择的特征个数由C决定
logistic = LogisticRegression(penalty='l2',C=1)
SelectFromModel(logistic).fit_transform(data,target)
```

```python
from sklearn.ensemble import GradientBoostingClassifier

#GBDT作为基模型的特征选择
gbdt = GradientBoostingClassifier()
SelectFromModel(gbdt).fit_transform(data,target)
```
