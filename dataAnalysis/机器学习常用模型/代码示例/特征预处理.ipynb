```python
from sklearn.datasets import load_iris
from sklearn.preprocessing import MinMaxScaler,StandardScaler,Normalizer
```

### 无量纲化处理

消除因为样本特征之间量级不同而导致的算法不精确问题

- 1.有些具有伸缩不变性的算法，可以不做无量纲处理
- 2.但是无量纲处理一般来说，不论任何算法都可以做，有利而无害

```python
data = load_iris().data
target = load_iris().target
```

```python
# 区缩放法
MinMaxScaler().fit_transform(data)
```

```python
# 标准化
StandardScaler().fit_transform(data)
```

```python
# 归一化
Normalizer().fit_transform(data)
```

```python
### 二值化处理
from sklearn.preprocessing import Binarizer
import numpy as np

train = np.array([[10,11,15,9,6,7,9,8]])

Binarizer(threshold=10).fit_transform(train)
```

```python
from numpy import vstack, array, nan
from sklearn.preprocessing import Imputer

#缺失值计算，返回值为计算缺失值后的数据
#参数missing_value为缺失值的表示形式，默认为NaN
#参数strategy为缺失值填充方式，默认为mean（均值）
train = vstack((array([nan, nan, nan, nan]), data))
# display(train)
# 有三种值可以填充strategy
# 平均值 mean
# 中值 median
# 众数 most_frequent
Imputer(strategy='mean').fit_transform(train)
```

```python
import pandas as pd
from pandas import Series,DataFrame

DataFrame(train).fillna('1')
```

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
# 哑编码
train = np.array([['male','yes'],
                  ['femal','no'],
                  ['male','yes'],
                  ['femal','yes'],
                  ['femal','yes'],
                  ['femal','no'],
                  ['male','no'],
                  ['male','yes'],
                  ['femal','no'],
                  ['male','no'],])
# 先使用labelencoder转换成数字
lable_enc1 = LabelEncoder()
enc1 = lable_enc1.fit_transform(train[:,0])
# 独热编码，把数字转换成矩阵，消除特征间的关联性
onehot_enc1 = OneHotEncoder()
# 训练的数据必须是列向量，跟机器学习模型是一致的
matrix1 = onehot_enc1.fit_transform(enc1.reshape(-1,1))
# 提取矩阵结果todense()
matrix1.todense()

lable_enc2 = LabelEncoder()
enc2 = lable_enc2.fit_transform(train[:,1])
onehot_enc2 = OneHotEncoder()
matrix2 = onehot_enc2.fit_transform(enc2.reshape(-1,1))
matrix2.todense()

n_train = np.concatenate((matrix1.todense(),matrix2.todense()),axis=1)
n_train
```
